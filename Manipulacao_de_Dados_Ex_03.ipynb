#### Tarefa 1. Importe o dataset e salve os dados em um dataframe

Os dados estão salvos no arquivo ***datasets/users_dataset.txt***.\
Salve os dados em uma variável de nome *users*.

````python
import pandas as pd
users = pd.read_csv('datasets/users_dataset.csv', sep = '|')
users.head()
````

#### Tarefa 2. Descubra qual a idade média por ocupação.

*Dica: use a função groupby do Pandas na coluna occupation*

````python
users.groupby(['occupation'])['age'].mean().round(2)
````

#### Tarefa 3. Para cada ocupação descubra a idade mínima e máxima.

````python
max_ = users.groupby(['occupation'])['age'].agg(age = max)
min_ = users.groupby(['occupation'])['age'].agg(age = min)

idades = max_
idades.rename(columns={'age' : 'Idade máxima'}, inplace=True)
idades['Idade mínima'] = min_
idades
````

#### Tarefa 4. Para cada combinação de ocupação e sexo, calcule a idade média.

Exemplo:

```
occupation     gender
administrator  F         40.638889
               M         37.162791
artist         F         30.307692
               M         32.333333
```

````python
users.groupby(['occupation', 'gender'])['age'].agg(age = 'mean')
````

#### Tarefa 5. Para cada ocupação calcule a porcentagem de homens e mulheres.

Exemplo:

```
occupation     gender
administrator  F          45.569620
               M          54.430380
artist         F          46.428571
               M          53.571429
doctor         M         100.000000
educator       F          27.368421
               M          72.631579
```

````python
genero = users.groupby(['occupation', 'gender'])['gender'].agg(gender = 'count')
genero_t = users.groupby(['occupation'])['occupation'].agg(occupation = 'count')
genero['porcentagem'] = ((genero.gender / genero_t.occupation) * 100).round(2)
genero
````
